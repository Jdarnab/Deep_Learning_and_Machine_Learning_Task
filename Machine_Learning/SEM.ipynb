```python
from scipy.stats import zscore

mapping = {
  "Never":1, "Rarely":2, "Occasionally":3, "Often":4, "Always":5,
  "Very Poor":1, "Poor":2, "Satisfactory":3, "Good":4, "Very Good":5,
  "18-25 years old":2, "41-65 years old":4, "More than 65 years old":5, "26-40 years old":3, "Less than 18 years old":1,
  "University level and above":3, "Secondary school level":2, "Primary school level and below":1,
  "7:10AM to 11:00AM (Morning peak hour)":1, "11:01 AM to 04:00PM (Off peak hour)":2, "04:01PM to 09:40PM (Afternoon peak hour)":3,
}


for col in df.columns:
    df[col]=df[col].replace(mapping)

for col in df_male.columns:
  df_male[col]=df_male[col].replace(mapping)

for col in df_female.columns:
  df_female[col]=df_female[col].replace(mapping)
```

```python
# quick standardization (optional but helps numeric stability)

ped_cols = ['Overall Safety','Overall security condition on footpath', 'Overall Satisfaction']
ped_data = df[ped_cols].apply(zscore)

# compute Cronbach's alpha
k = ped_data.shape[1]
item_variances = ped_data.var(ddof=1)
total_var = ped_data.sum(axis=1).var(ddof=1)
alpha = (k / (k - 1)) * (1 - item_variances.sum() / total_var)
print(f"Cronbach's alpha for PedestrianSafety indicators = {alpha:.3f}")
```

```python
!pip install factor_analyzer
from factor_analyzer.factor_analyzer import calculate_kmo, calculate_bartlett_sphericity

# exclude your 3 target indicators for now
X = df.drop(columns=ped_cols, errors='ignore').select_dtypes(include=['float','int'])

# KMO (sampling adequacy)
kmo_all, kmo_model = calculate_kmo(X)

# Bartlett’s test (sphericity)
chi_square_value, p_value = calculate_bartlett_sphericity(X)

print(f"KMO overall measure = {kmo_model:.3f}")
print(f"Bartlett’s test: chi2 = {chi_square_value:.3f}, p = {p_value:.3f}")
```

```python
from factor_analyzer import FactorAnalyzer, Rotator
from sklearn.preprocessing import StandardScaler

# select numeric non-target variables
X = df.drop(columns=ped_cols, errors='ignore').select_dtypes(include=['float','int'])
X_scaled = StandardScaler().fit_transform(X)

# run EFA with 5 factors (as in TRB paper)
fa = FactorAnalyzer(n_factors=5, rotation=None, method='principal')
fa.fit(X_scaled)

# Varimax rotation
rotator = Rotator(method='varimax')
rotated_loadings = rotator.fit_transform(fa.loadings_)

# show rotated loadings
rot_df = pd.DataFrame(rotated_loadings, index=X.columns, columns=[f"Factor{i+1}" for i in range(5)])
print(rot_df.round(3))
```

```python
from semopy import Model

model_desc = """
# Measurement model (latent = ~ observed indicators)
Pedestrian_Infrastructure_Quality =~ Shaded_Facilities + Road_Marking_Signage + Bike_Parking + Inclusive_Design
Risky_Pedestrian_Behavior =~ Undesignated_Crossing + Hand_Gesture_Crossing + Transit_Rush + Distracted_Walking + Group_Crossing_Behavior + Group_Footpath_Avoidance
Crossing_and_Footpath_Safety =~ Footpath_Condition + Crossing_Facilities + Street_Lighting
Unsafe_Road_Usage =~ Road_Walking + Barrier_Jumping + Reckless_Crossing
Footpath_Obstruction_and_Maintenance =~ Footpath_Maintenance + Footpath_Obstruction_Control

# Target latent (endogenous)
PedestrianSafety =~ Overall_Safety + Overall_Footpath_Security + Overall_Satisfaction

# --- Structural model (regression paths)
PedestrianSafety ~ Pedestrian_Infrastructure_Quality + Risky_Pedestrian_Behavior + Crossing_and_Footpath_Safety + Unsafe_Road_Usage + Footpath_Obstruction_and_Maintenance

# Allow latent correlations (optional)
Pedestrian_Infrastructure_Quality ~~ Risky_Pedestrian_Behavior
Pedestrian_Infrastructure_Quality ~~ Crossing_and_Footpath_Safety
Pedestrian_Infrastructure_Quality ~~ Unsafe_Road_Usage
Pedestrian_Infrastructure_Quality ~~ Footpath_Obstruction_and_Maintenance
Risky_Pedestrian_Behavior ~~ Crossing_and_Footpath_Safety
Risky_Pedestrian_Behavior ~~ Unsafe_Road_Usage
Risky_Pedestrian_Behavior ~~ Footpath_Obstruction_and_Maintenance
Crossing_and_Footpath_Safety ~~ Unsafe_Road_Usage
Crossing_and_Footpath_Safety ~~ Footpath_Obstruction_and_Maintenance
Unsafe_Road_Usage ~~ Footpath_Obstruction_and_Maintenance
"""
```

```python
model = Model(model_desc)
model.fit(df)

# View estimates
est = model.inspect(std_est= True)
print(est.head(10))

# Model fit stats
from semopy import calc_stats
stats = calc_stats(model)
print(stats)
```

```python
from graphviz import Digraph

# Initialize Graphviz diagram
dot = Digraph(format='png')
dot.attr(rankdir='LR')  # Left-to-right layout

#  Define latent variables and their indicators
latent_structure = {
    "Pedestrian_Infrastructure_Quality": [
        "Shaded_Facilities", "Road_Marking_Signage", "Bike_Parking", "Inclusive_Design"
    ],
    "Risky_Pedestrian_Behavior": [
        "Undesignated_Crossing", "Hand_Gesture_Crossing", "Transit_Rush",
        "Distracted_Walking", "Group_Crossing_Behavior", "Group_Footpath_Avoidance"
    ],
    "Crossing_and_Footpath_Safety": [
        "Footpath_Condition", "Crossing_Facilities", "Street_Lighting"
    ],
    "Unsafe_Road_Usage": [
        "Road_Walking", "Barrier_Jumping", "Reckless_Crossing"
    ],
    "Footpath_Obstruction_and_Maintenance": [
        "Footpath_Maintenance", "Footpath_Obstruction_Control"
    ],
    "PedestrianSafety": [
        "Overall_Safety", "Overall_Footpath_Security", "Overall_Satisfaction"
    ]
}

# Add latent variables as ellipses
for latent in latent_structure:
    dot.node(latent, shape='ellipse')

# Add observed indicators as rectangles and connect them
for latent, indicators in latent_structure.items():
    for ind in indicators:
        dot.node(ind, shape='box')
        dot.edge(latent, ind)

# Structural regression paths (endogenous model)
dot.edge("Pedestrian_Infrastructure_Quality", "PedestrianSafety")
dot.edge("Risky_Pedestrian_Behavior", "PedestrianSafety")
dot.edge("Crossing_and_Footpath_Safety", "PedestrianSafety")
dot.edge("Unsafe_Road_Usage", "PedestrianSafety")
dot.edge("Footpath_Obstruction_and_Maintenance", "PedestrianSafety")

# Latent correlations (bidirectional edges)
correlated_latents = [
    ("Pedestrian_Infrastructure_Quality", "Risky_Pedestrian_Behavior"),
    ("Pedestrian_Infrastructure_Quality", "Crossing_and_Footpath_Safety"),
    ("Pedestrian_Infrastructure_Quality", "Unsafe_Road_Usage"),
    ("Pedestrian_Infrastructure_Quality", "Footpath_Obstruction_and_Maintenance"),
    ("Risky_Pedestrian_Behavior", "Crossing_and_Footpath_Safety"),
    ("Risky_Pedestrian_Behavior", "Unsafe_Road_Usage"),
    ("Risky_Pedestrian_Behavior", "Footpath_Obstruction_and_Maintenance"),
    ("Crossing_and_Footpath_Safety", "Unsafe_Road_Usage"),
    ("Crossing_and_Footpath_Safety", "Footpath_Obstruction_and_Maintenance"),
    ("Unsafe_Road_Usage", "Footpath_Obstruction_and_Maintenance")
]

for a, b in correlated_latents:
    dot.edge(a, b, dir='both', style='dashed', color='gray')

# Render the diagram
dot.render("pedestrian_sem_diagram", view=True)
```
